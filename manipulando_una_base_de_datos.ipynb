```python
#Creando la base de datos local con la biblioteca SQLAlchemy.
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text
engine = create_engine('sqlite:///:memory:')
import pandas as pd
```

```python
#Escribiendo los datos del archivo CSV en esta base de datos local.
archivo = '/content/drive/MyDrive/ETL/PANDAS/PANDAS I O/clientes_banco.csv'
datos = pd.read_csv(archivo)
datos.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
query = 'SELECT * FROM Clientes'
pd.read_sql(sql=text(query), con=engine.connect()) #con = conexión
```

```python
#Actualizar el registro del cliente ID 6840104 cuyo rendimiento anual cambió a 300000.
#Actualizando un registro de base de datos.
#Cuando ejecutemos una actualización de un banco de datos, puede que funcione o no funcione.
#Para identificar que tipo de error estará sucediendo en el futuro, se hace lo siguiente
from sqlalchemy.exc import SQLAlchemyError
query = "UPDATE clientes SET Rendimiento_anual=300000 WHERE ID_Cliente=6840104"
try: #intenta ejecutar ese sql, y si no se puede, pues se activa except
  r_set = engine.connect().execute(text(query)) #primero se ejecuta el motor, y luego se muestra alguna cosa
except SQLAlchemyError as e:
  print(e) #muestra el detalle de ese error
else:
  print('#Registros actualizados: ', r_set.rowcount) #muestra el numero de registros actualizados
```

```python
#Eliminar el registro de cliente ID 5008809, ya que esta persona ya no tiene cuenta en la institución financiera.
query = "DELETE FROM clientes WHERE ID_Cliente=5008809"
try: #intenta ejecutar ese sql, y si no se puede, pues se activa except
  r_set = engine.connect().execute(text(query)) #primero se ejecuta el motor, y luego se muestra alguna cosa
except SQLAlchemyError as e:
  print(e) #muestra el detalle de ese error
else:
  print('#Numero de registros borrados: ', r_set.rowcount) #muestra el numero de registros borrados
```
